## Teste de Hipótese

#### Hipótese nula -> A distribuição da taxa da gorjeta é a mesa nos dois grupos

#### Hipótese alternativa -> A distribuição da taxa da gorjeta não é a mesma nos dois grupos

```python
from scipy.stats import ranksums
```

```python
sobremesa = novo_df.query("Sobremesa == 'Sim'").Porcentagem
```

```python
sem_sobremesa = novo_df.query("Sobremesa == 'Não'").Porcentagem
```

```python
r = ranksums(sobremesa, sem_sobremesa)
```

```python
print(f'O valor do pvalue é {r.pvalue}.')
```

#### Hipótese nula -> A distribuição da taxa da gorjeta é a mesa nos dois grupos

## Teste de Hipótese

#### Hipótese nula -> A distribuição do valor da conta é igual no sábado e no domingo

#### Hipóstese alternativa -> A distribuição do valor da conta não é igual no sábado e no domingo

```python
conta_domingo = novo_df.query("Dia_da_semana == 'Domingo'").Valor_da_conta
```

```python
conta_sabado = novo_df.query("Dia_da_semana == 'Sábado'").Valor_da_conta
```

```python
r1 = ranksums(conta_domingo, conta_sabado)
```

```python
print(f'O valor do pvalue é {r1.pvalue}.')
```

#### Hipótese nula -> A distribuição do valor da conta é igual no sábado e no domingo

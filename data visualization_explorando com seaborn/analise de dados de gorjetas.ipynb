# Análise de Dados

## Importando base de dados

```python
import pandas as pd
```

```python
dados = pd.read_csv('arquivos/tips.csv')
dados
```

## Renomeando os campos

```python
dados.columns
```

```python
renomear = {'total_bill': 'Valor_da_conta', 
            'tip': 'Gorjeta', 
            'dessert': 'Sobremesa', 
            'day': 'Dia_da_semana', 
            'time': 'Hora_do_dia', 
            'size': 'Total_de_pessoas'}
```

```python
novo_df = dados.rename(columns = renomear)
```

```python
novo_df
```

```python
novo_df.Sobremesa.unique()
```

```python
sim_nao = {'No': 'Não', 'Yes': 'Sim'}
```

```python
novo_df.Sobremesa.map(sim_nao)
```

```python
novo_df.Sobremesa = novo_df.Sobremesa.map(sim_nao)
```

```python
novo_df
```

```python
novo_df.Dia_da_semana.unique()
```

```python
dias = {'Sun': 'Domingo', 'Sat': 'Sábado', 'Thur': 'Quinta-feira', 'Fri': 'Sexta-feira'}
```

```python
novo_df.Dia_da_semana = novo_df.Dia_da_semana.map(dias)
```

```python
novo_df
```

```python
novo_df.Hora_do_dia.unique()
```

```python
hora = {'Dinner': 'Jantar', 'Lunch': 'Almoço'}
```

```python
novo_df.Hora_do_dia = novo_df.Hora_do_dia.map(hora)
```

```python
novo_df
```

```python
novo_df.to_csv('arquivos/tips_tratada.csv', index = False)
```

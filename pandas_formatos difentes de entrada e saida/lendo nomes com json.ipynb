## Criando Nomes

```python
import pandas as pd
```

```python
nomes_m = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-m.json')
nomes_f = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-f.json')
```

```python
nomes_f
```

```python
nomes_m.sample(5)
```

```python
nomes_f.sample(5)
```

```python
print('Quantidade de nomes masculinos:' + str(len(nomes_m)) + ' Quantidades de nomes femininos:' + str(len(nomes_f)))
```

```python
frames = [nomes_m, nomes_f]
```

```python
nomes = pd.concat(frames)['nome'].to_frame()
```

```python
nomes
```

## Incluindo ID dos Alunos

```python
import numpy as np
```

```python
total_alunos = len(nomes)
total_alunos
```

```python
nomes['id_aluno'] = np.random.permutation(total_alunos) + 1
```

```python
nomes
```

```python
dominios = ['@dominiodoemail.com.br', '@serviçodoemail.com']
nomes['domínio'] = np.random.choice(dominios, total_alunos)
```

```python
nomes
```

```python
nomes['email'] = nomes.nome.str.cat(nomes.domínio).str.lower()
```
